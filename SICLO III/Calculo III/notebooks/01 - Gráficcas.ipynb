# Gráficas

```julia
using Plots
```

## 1. Gráficas en dos dimeciones

```julia
x = -pi:0.01:pi
y = sin.(x)

plot(x, y, label = "sin(x)")
```

## 2. Layout

```julia

plot(x, y, layout=(2,1), label = "sin(x)")
```

## 3. Plot3D

```julia
t = -pi:pi/200:4pi
x = cos.(t)
y = sin.(t)
plot3d(x, y, t, lw=pi/4)
```

## 4. Superficie

```julia
x = -1:0.1:1
y = -1:0.1:1
h(x, y) = x^2 - y^2

surface(x, y, h)
```

## 5. Malla

```julia
wireframe(x, y, h)
```
